# Predicting heart disease using machine learning

This project predicts whether or not someone has heart disease based on their medical attributes. 
We are using different Python-based machine learning libraries and data science tools to solve this problem.

We are going to take the following approach:
1. Problem definition
2. Data
3. Evaluation
4. Features
5. Modelling
6. Experimentation

## 1. Problem definition
In a statement,
> given clinical parameters about a patient, can we predict whether or not someone has heart disease

## 2. Data
The original data came from the Clevland data from the UCI Machine Learning repository.
There is also a version of the dataset available on kaggle.

## 3. Evaluation
> If we can reach 95% accuracy at predicting whether or not someone has heart disease during the proof of concept,
we'll pursue the project.

## 4. Features
This is where you will get information about the features of your data

** Create data dictionary **

* age: age in years
* sex: sex (1 = male; 0 = female)
* cp: chest pain type
-- Value 0: typical angina
-- Value 1: atypical angina
-- Value 2: non-anginal pain
-- Value 3: asymptomatic
* trestbps: resting blood pressure (in mm Hg on admission to the hospital)
* chol: serum cholestoral in mg/dl
* fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* restecg: resting electrocardiographic results
-- Value 0: normal
-- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
-- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
* thalach: maximum heart rate achieved
* exang: exercise induced angina (1 = yes; 0 = no)
* oldpeak = ST depression induced by exercise relative to rest
* slope: the slope of the peak exercise ST segment
-- Value 0: upsloping
-- Value 1: flat
-- Value 2: downsloping
* ca: number of major vessels (0-3) colored by flourosopy
* thal: 0 = normal; 1 = fixed defect; 2 = reversable defect
and the label
* condition: 0 = no disease, 1 = disease

# 5. Modelling

```python
df.head()
```

```python
# Splitting our data into features & labels
X = df.drop('target', axis=1)
y = df['target']
```

```python
X
```

```python
y
```

```python
# Splitting our data into train and test
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### Now that we have got our data split into training and test sets, it's time do build a new machine learning model

We will train it (find patterns) on a training data.

And we will test it (use the patterns) on a test data.

We will try 3 different classification models:

    1.LogisticRegression 
    2.K-Nearest Neighbors Classifier
    3.RandomForestClassifier

```python
# Create a dictionary of models
models = {'Logistic Regression': LogisticRegression(),
          'KNN': KNeighborsClassifier(),
          'Random Forest': RandomForestClassifier()}

def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Definition:
        Fits and scores the dictionary of models and returns the results in a new dictionary
    
    Arguments:
        models: dictionary of models
        X_train: training set (no labels)
        X_test: testing set (no labels)
        y_train: training set (labels)
        y_test: testing set (labels)
    """
    np.random.seed(42)
    model_scores = {}
    
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        model_scores[model_name] = model.score(X_test, y_test)
    
    return model_scores
```

```python
model_scores = fit_and_score(models, X_train, X_test, y_train, y_test)
model_scores
```

```python
model_compare = pd.DataFrame(model_scores, index=['accuracy']).T

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(model_compare.index, model_compare['accuracy'])
ax.set_title('Comparison of 3 different model\'s performance', pad=10)
ax.set_ylabel('Accuracy', labelpad=10);
```

### Now let's look at the following:
* Hyperparameter tuning
* Feature importance
* Confusion matrix
* Cross-validation

```python
# Let's tune KNN
train_scores = []
test_scores = []

neighbors = range(1, 21)

for i in neighbors:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    train_scores.append(knn.score(X_train, y_train))
    test_scores.append(knn.score(X_test, y_test))

train_scores, test_scores
```

```python
#Now let's visualize our scores

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(neighbors, train_scores, label='Train scores')
ax.plot(neighbors, test_scores, label='Test scores');
ax.set_title('KNN model train and test scores', pad=10)
ax.set_ylabel('Model score', labelpad=10)
ax.set_xlabel('n_neighbors', labelpad=10)
ax.legend()
plt.xticks(np.arange(1, 21, 1));

print(f'Maximum score of KNN model on testing data is {max(test_scores)*100:.2f}%')
```

 Even after hyperparameter tuning our KNN model's performance is inferior to other models

## Hyperparameter tuning using RandomizedSearchCV
We're going to tune:
* LogisticRegression
* RandomForestClassifier

```python
#Create a hyperparameter grids for LogisticRegression and RandomForestClassifier
log_reg_grid = {'C': np.logspace(-4, 4, 20), 'solver': ['liblinear']}
rf_grid = {
    'n_estimators': np.arange(10, 1000, 50),
    'max_depth': [None, 3, 5, 10],
    'min_samples_split': np.arange(2, 20, 2),
    'min_samples_leaf': np.arange(1, 20, 2)
}
```

Now that we have both of our grids, let's tune them using RandomizedSearchCV

```python
# Tune LogisticRegression
np.random.seed(42)

rs_log_reg = RandomizedSearchCV(LogisticRegression(), log_reg_grid, n_iter=20, cv=5, verbose=True)
rs_log_reg.fit(X_train, y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(X_test, y_test)
```

```python
#After tuning the score remains the same
model_scores
```

```python
#Now let's tune the RandomForestClassifier

np.random.seed(42)
rs_rf = RandomizedSearchCV(RandomForestClassifier(), rf_grid, n_iter=20, cv=5, verbose=True)
rs_rf.fit(X_train, y_train)
```

```python
rs_rf.best_params_
```

```python
rs_rf.score(X_test, y_test)
```

```python
# There is a slight improvement in the score but it's still under LogisticRegression
model_scores
```

```python
# As LogisticRegression performs the best let's tune its hyperparameters using GridSearchCV
np.random.seed(42)
log_reg_grid = {'C': np.logspace(-4, 4, 30), 'solver': ['liblinear']}
gs_log_reg = GridSearchCV(LogisticRegression(), log_reg_grid, cv=5, verbose=True)
gs_log_reg.fit(X_train, y_train)
```

```python
gs_log_reg.best_params_
```

```python
gs_log_reg.score(X_test, y_test)
```

```python
# Score remains the same
model_scores
```

```python
# To evaluate our model first we need to make predictions
y_preds = gs_log_reg.predict(X_test)
y_preds_proba_positive = gs_log_reg.predict_proba(X_test)[:, 1]
```

```python
fpr, tpr, _ = roc_curve(y_test, y_preds_proba_positive)
roc_auc = roc_auc_score(y_test, y_preds_proba_positive)

display_roc = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='LogisticRegression')
display_roc.plot()
plt.title('Receiver operating characteristic curve', pad=10)
plt.ylabel('True Positive Rate', labelpad=10)
plt.xlabel('False Positive Rate', labelpad=10);
```

```python
# Ok, now let's checkout our confusion matrix
confusion_matrix(y_test, y_preds)
```

```python
# Let's visualize it!
sns.set(font_scale=1.2)

def plot_conf_matrix(y_test, y_preds):
    """
    Plots a nice looking confusion matrix using Seaborn's heatmap
    """
    fig, ax = plt.subplots(figsize=(5, 3))
    ax = sns.heatmap(confusion_matrix(y_test, y_preds), annot=True, cmap='inferno')
    ax.set_ylabel('True label', labelpad=10)
    ax.set_xlabel('Predicted label', labelpad=10)
    
plot_conf_matrix(y_test, y_preds);
```

## Let's calculate Accuracy, Precision, Recall and F1 score using cross-validation

```python
gs_log_reg.best_params_
```

```python
clf = LogisticRegression(C=0.20433597178569418, solver='liblinear')
```

```python
# Let's write a function that calculates our parameters
def cv_evaluate_metrics(clf, X, y, cv=5):
    cv_accuracy = np.mean(cross_val_score(clf, X=X, y=y, cv=cv, scoring='accuracy'))
    cv_precision = np.mean(cross_val_score(clf, X=X, y=y, cv=cv, scoring='precision'))
    cv_recall = np.mean(cross_val_score(clf, X=X, y=y, cv=cv, scoring='recall'))
    cv_f1 = np.mean(cross_val_score(clf, X=X, y=y, cv=cv, scoring='f1'))
    
    eval_metrics_dict = {
        'Accuracy': cv_accuracy,
        'Precision': cv_precision,
        'Recall': cv_recall,
        'F1': cv_f1
    }
    
    return eval_metrics_dict
```

```python
eval_metrics_dict = cv_evaluate_metrics(clf, X, y)
```

```python
# Now let's visualize our evaluation metrics
cv_metrics = pd.DataFrame(eval_metrics_dict, index=['score']).T
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(cv_metrics.index, cv_metrics['score'], color='salmon', edgecolor='black', linewidth=0.2)
ax.set_title('Comparison of evaluation metrics', pad=10)
ax.set_ylabel('Score', labelpad=10);
```
